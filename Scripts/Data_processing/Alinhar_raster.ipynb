```python
import os
import rasterio
import numpy as np
from rasterio.warp import reproject, Resampling
from collections import Counter
```

```python
# ================================
# CONFIGURA√á√ïES
# ================================
folder = "/content/drive/MyDrive/modelo/Regiao_1"     # onde est√£o os rasters
output_folder = "/content/drive/MyDrive/ALINHADOS"
os.makedirs(output_folder, exist_ok=True)
```

```python
# ================================
# 1) COLETAR INFORMA√á√ïES
# ================================
tifs = [os.path.join(folder, f) for f in os.listdir(folder) if f.endswith(".tif")]

sizes = []
meta_info = {}

for tif in tifs:
    with rasterio.open(tif) as src:
        size = (src.height, src.width)
        sizes.append(size)
        meta_info[tif] = {
            "size": size,
            "transform": src.transform,
            "crs": src.crs,
            "dtype": src.dtypes[0]
        }

# tamanho mais comum
common_size = Counter(sizes).most_common(1)[0][0]
print("üìå Tamanho base escolhido:", common_size)
```

```python
# ================================
# 2) DEFINIR RASTER BASE AUTOM√ÅTICO
#    (o primeiro raster que tem o tamanho mais comum)
# ================================
for tif in tifs:
    if meta_info[tif]["size"] == common_size:
        raster_base = tif
        break

print("üìå Raster base:", os.path.basename(raster_base))

# carregar info do base
with rasterio.open(raster_base) as ref:
    ref_profile = ref.profile
    ref_transform = ref.transform
    ref_crs = ref.crs
    ref_shape = (ref.height, ref.width)
```

```python
# ================================
# 3) IDENTIFICAR QUAIS EST√ÉO DIFERENTES EM QUALQUER ASPECTO
# ================================

diferentes = []

with rasterio.open(raster_base) as ref:
    base_size = (ref.height, ref.width)
    base_res = ref.res
    base_transform = ref.transform
    base_crs = ref.crs

print("\nüìå Raster base:", os.path.basename(raster_base))
print("  - Tamanho:", base_size)
print("  - Resolu√ß√£o:", base_res)
print("  - CRS:", base_crs)
print("  - Transform:", base_transform, "\n")

print("========================================")
print("üìå RASTERS DIFERENTES EM QUALQUER CRIT√âRIO:")
print("========================================")

for tif in tifs:
    with rasterio.open(tif) as src:
        size_diff = (src.height, src.width) != base_size
        res_diff  = src.res != base_res
        crs_diff  = src.crs != base_crs
        trans_diff = src.transform != base_transform

        if size_diff or res_diff or crs_diff or trans_diff:
            diferentes.append(tif)
            print(f" - {os.path.basename(tif)}")

            # Mostrar motivo(s)
            if size_diff:
                print("     ‚Ä¢ Tamanho diferente:", (src.height, src.width))
            if res_diff:
                print("     ‚Ä¢ Resolu√ß√£o diferente:", src.res)
            if crs_diff:
                print("     ‚Ä¢ CRS diferente:", src.crs)
            if trans_diff:
                print("     ‚Ä¢ Origem/Transform diferente:", src.transform)

print("\nTotal encontrados:", len(diferentes), "rasters diferentes.\n")

if len(diferentes) == 0:
    print("üéâ Todos os rasters s√£o id√™nticos ao base. Nada a alinhar.\n")
```

```python
# ================================
# 4) ALINHAR SOMENTE OS DIFERENTES
# ================================
from rasterio.warp import reproject, Resampling
import numpy as np
import os
import rasterio

if len(diferentes) == 0:
    print("üéâ Todos os rasters j√° est√£o alinhados! Nada para fazer.")
else:
    print("\nüîß Iniciando alinhamento dos rasters diferentes...\n")

    for tif in diferentes:
        nome = os.path.basename(tif)
        out_tif = os.path.join(output_folder, nome)

        with rasterio.open(tif) as src:
            data = src.read(1)
            nodata_src = src.nodata  # ‚¨Ö pega o NoData ORIGINAL

            # matriz preenchida COM O NODATA ORIGINAL
            aligned = np.full(ref_shape, nodata_src, dtype=data.dtype)

            # reproje√ß√£o preservando NoData
            reproject(
                source=data,
                destination=aligned,
                src_transform=src.transform,
                src_crs=src.crs,
                dst_transform=ref_transform,
                dst_crs=ref_crs,
                src_nodata=nodata_src,   # ‚¨Ö diz ao reproject qual √© NoData
                dst_nodata=nodata_src,   # ‚¨Ö mant√©m o mesmo NoData
                resampling=Resampling.nearest
            )

            # cria perfil com o MESMO NoData do raster original
            profile = ref_profile.copy()
            profile.update(
                dtype=data.dtype,
                nodata=nodata_src        # ‚¨Ö preservado!
            )

            # salva resultado
            with rasterio.open(out_tif, "w", **profile) as dst:
                dst.write(aligned, 1)

        print(f"‚úî Alinhado: {nome} (NoData preservado)")

    print("\nüéâ TODOS OS RASTERS DIFERENTES FORAM ALINHADOS SEM ALTERAR O NODATA!\n")
```

```python
# =====================================================
# VERIFICAR RESOLU√á√ÉO E IDENTIFICAR RASTERS DIFERENTES
# =====================================================

# Pega a resolu√ß√£o do raster-base
with rasterio.open(raster_base) as ref:
    res_base = ref.res
    print("üìå Resolu√ß√£o base:", res_base)

# Lista rasters com resolu√ß√£o diferente
res_diferentes = []

for tif in tifs:
    with rasterio.open(tif) as src:
        if src.res != res_base:
            res_diferentes.append(tif)

print("\n========================================")
print("üìå RASTERS COM RESOLU√á√ÉO DIFERENTE:")
print("========================================")

if len(res_diferentes) == 0:
    print("üéâ Todos os rasters t√™m a mesma resolu√ß√£o! Nada para alinhar.")
else:
    for r in res_diferentes:
        print(" -", os.path.basename(r))

print("\nTotal:", len(res_diferentes), "rasters com resolu√ß√£o diferente\n")

# =====================================================
# ALINHAR SOMENTE OS RASTERS COM RESOLU√á√ÉO DIFERENTE
# =====================================================

for tif in diferentes:
    nome = os.path.basename(tif)
    out_tif = os.path.join(output_folder, nome)

    with rasterio.open(tif) as src:
        data = src.read(1)

        # preservar NoData original
        orig_nodata = src.nodata

        # matriz alinhada
        aligned = np.zeros(ref_shape, dtype=data.dtype)

        reproject(
            source=data,
            destination=aligned,
            src_transform=src.transform,
            src_crs=src.crs,
            dst_transform=ref_transform,
            dst_crs=ref_crs,
            resampling=Resampling.nearest
        )

        # manter nodata original
        profile = ref_profile.copy()
        profile.update(
            dtype=data.dtype,
            nodata=orig_nodata
        )

        with rasterio.open(out_tif, "w", **profile) as dst:
            dst.write(aligned, 1)

    print(f"‚úî Alinhado: {nome}")

print("\nüéâ Todos os raster diferentes foram alinhados sem alterar NoData!")
```

```python
# ================================
# 3) ALINHAR TODOS
# ================================
for tif in tifs:
    nome = os.path.basename(tif)
    out_tif = os.path.join(output_folder, nome)

    with rasterio.open(tif) as src:
        data = src.read(1)

        # inicia com zeros do tamanho certo
        aligned = np.zeros(ref_shape, dtype=data.dtype)

        # reprojetar/alinhar
        reproject(
            source=data,
            destination=aligned,
            src_transform=src.transform,
            src_crs=src.crs,
            dst_transform=ref_transform,
            dst_crs=ref_crs,
            resampling=Resampling.nearest  # para classes
        )

        # atualizar metadados
        profile = ref_profile.copy()
        profile.update(dtype=data.dtype)

        with rasterio.open(out_tif, "w", **profile) as dst:
            dst.write(aligned, 1)

    print(f"‚úî Alinhado: {nome}")

print("\nüéâ TODOS OS RASTERS FORAM ALINHADOS COM SUCESSO!")
```

```python
# 1. Instalar as bibliotecas (se necess√°rio)
!pip install geopandas rasterio numpy

import rasterio
from rasterio.features import rasterize
import geopandas as gpd
import numpy as np
import os
from google.colab import drive

# 2. Montar o Google Drive
drive.mount('/content/gdrive')

# --- CONFIGURA√á√ÉO DOS ARQUIVOS E CAMINHOS PARA A REGI√ÉO 24 ---
BASE_DIR_24 = "/content/gdrive/MyDrive/modelo/Regiao_24/"

# Raster principal para alinhamento (Mude o nome do Mapbiomas se for diferente)
path_mapbiomas_24 = os.path.join(BASE_DIR_24, "mapbiomas_2018_24.tif")

# Shapefile de limite original
path_regiao_original_24 = os.path.join(BASE_DIR_24, "regiao_24.shp")

# Arquivo TIF de M√°scara de Sa√≠da (ESTE √â O ARQUIVO PARA O DINAMICA EGO)
path_mascara_final_24 = os.path.join(BASE_DIR_24, "Mascara_Alocacao_FINAL_Regiao24.tif")

# Dist√¢ncia do buffer (1 pixel * 30.0m)
BUFFER_DISTANCE_METERS = 30.0

def criar_e_rasterizar_mascara_alinhada(path_dados, path_shp_in, path_tif_out, buffer_dist):
    print(f"Iniciando cria√ß√£o da M√°scara de Aloca√ß√£o para a Regi√£o 24...")

    # 1. Carregar o Shapefile Original
    try:
        limite_gdf = gpd.read_file(path_shp_in)
    except Exception as e:
        print(f"ERRO ao carregar o Shapefile original: {e}")
        return

    # 2. Criar o Buffer (1 pixel / 30 metros)
    print(f"Aplicando buffer de {buffer_dist} metros...")
    # O buffer() opera no CRS original, que est√° em metros (confirmado por voc√™).
    limite_buffer_gdf = limite_gdf.buffer(buffer_dist)
    buffer_gdf = gpd.GeoDataFrame(geometry=limite_buffer_gdf, crs=limite_gdf.crs)

    # 3. Carregar o Raster Mapbiomas para obter os metadados de alinhamento
    with rasterio.open(path_dados) as src_dados:
        transform = src_dados.transform
        out_shape = src_dados.shape
        out_crs = src_dados.crs

        # 4. Rasterizar o Buffer (usando o alinhamento do Mapbiomas)
        print("Rasterizando o Buffer para criar a M√°scara TIF...")
        try:
            mascara_raster = rasterize(
                [(geom, 1) for geom in buffer_gdf.geometry], # 1 dentro do buffer, 0 fora
                out_shape=out_shape,
                transform=transform,
                fill=0, # Preenche com 0 (fora do buffer)
                dtype=np.uint8
            )
        except Exception as e:
             print(f"ERRO durante a rasteriza√ß√£o: {e}.")
             return

        # 5. Salvar o Raster de M√°scara Final
        profile = src_dados.profile
        profile.update(dtype=rasterio.uint8, count=1, nodata=0) # M√°scara bin√°ria

        with rasterio.open(path_tif_out, 'w', **profile) as dst:
            dst.write(mascara_raster, 1)

    print(f"\n‚úÖ SUCESSO! O Mapa de M√°scara Final foi salvo em:")
    print(f"   -> {path_tif_out}")

# Executar a fun√ß√£o
criar_e_rasterizar_mascara_alinhada(
    path_mapbiomas_24,
    path_regiao_original_24,
    path_mascara_final_24,
    BUFFER_DISTANCE_METERS
)
```

```python
import rasterio
from rasterio.warp import calculate_default_transform, reproject, Resampling
import numpy as np
```

```python
crs_alvo = "ESRI:102015"

```

```python
raster_in = "/content/drive/MyDrive/modelo/Regiao_1/Precipitacao_1.tif"
raster_reproj = "/content/drive/MyDrive/modelo/Regiao_1/Precipitacao_1_reprojet.tif"

with rasterio.open(raster_in) as src:
    transform, width, height = calculate_default_transform(
        src.crs,
        crs_alvo,
        src.width,
        src.height,
        *src.bounds
    )

    meta = src.meta.copy()
    meta.update({
        "crs": crs_alvo,
        "transform": transform,
        "width": width,
        "height": height
    })

    with rasterio.open(raster_reproj, "w", **meta) as dst:
        for i in range(1, src.count + 1):
            reproject(
                source=rasterio.band(src, i),
                destination=rasterio.band(dst, i),
                src_transform=src.transform,
                src_crs=src.crs,
                dst_transform=transform,
                dst_crs=crs_alvo,
                resampling=Resampling.bilinear  # cont√≠nua
                # use Resampling.nearest para categ√≥rica
            )
```
