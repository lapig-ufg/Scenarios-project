```python
src = "/content/gdrive/MyDrive/TNC/AMZCERR_2023_final_30m_1.tif"
with rioxarray.open_rasterio(src) as src_2023:#rasterio.open(src) as src_2023:
    lulc_2023 = src_2023.sel(band=1)
```

```python
# Create a binary raster of transition: from class 15 to class 1
from_class = 15
to_class = 1
trans_cells = ((lulc_2018.values == 15) & (lulc_2023.values == 1)).astype(np.uint8)

import numpy as np
from scipy.ndimage import label

# Assume TRANS_CELLS is already a binary matrix (np.ndarray)
# with 1 for cells that changed class (e.g., from 15 to 1)

# 1. Label connected patches
structure = np.ones((3, 3), dtype=int)  # 8-connectivity
labeled, num_patches = label(trans_cells, structure=structure)

# 2. Calculate area (number of cells) per patch
areas = np.bincount(labeled.ravel())[1:]  # ignore label 0 (background)

# 3. Compute mean and standard deviation
area_mean = np.mean(areas)
area_sd = np.std(areas)

print(f"üìê Mean patch area: {area_mean:.2f} cells")
print(f"üìä Patch area standard deviation: {area_sd:.2f}")

```
